##### Here's what these certificates mean:

- **R**: This one stands for "Really, really not for kids." Expect lots of violence, sex, and bad language. And definitely don't bring your grandma to see this one.

- **PG**: This means "Pretty good for kids." There might be some mild language or violence, but overall it's safe for little eyes.

- **PG-13**: "Parents strongly cautioned to bring their teenagers." This one is for the older kids who can handle a bit more action and intensity.

- **Not Rated**: This is code for "We have no idea what's in this movie." It's a wild card, so buckle up and hope for the best.

- **G**: "Good for everyone." This one is safe for even the youngest movie-goers. Expect plenty of singing animals and happy endings.

- **Passed**: This one is old school. It means the movie was approved by censors who thought it was "Passable for public viewing."

- **Approved**: Another oldie but a goodie. This one means the movie was deemed "Acceptable for all audiences" by the censors.

- **18+**: This one is not for the faint of heart. It's like an R-rated movie, but even more intense. Leave the kids at home and prepare to cover your eyes.

- **Not Available**: This certificate means "You'll have to watch it and find out." It's a mystery!

- **TV-PG**: "Pretty good for TV." It's like the PG movie certificate, but for TV shows. Perfect for family-friendly sitcoms.

- **Unrated**: This means the movie hasn't been rated by the censors yet. Watch at your own risk!

- **X**: This is code for "XXX-rated." Expect plenty of sex and nudity. Definitely not a first date movie.

- **13+**: This one is for the tweens and teens. It's like PG-13, but for TV shows.

- **TV-MA**: "Mature audiences only." This one is like an R-rated movie, but for TV shows. Expect plenty of violence, sex, and bad language.

- **GP**: This stands for "General audiences, parental guidance suggested." It's like PG, but the parents need to be a bit more involved in deciding if it's right for their kids.

#### Content based filtering
Alright, let me try to explain content-based filtering in a way that won't put you to sleep!

So, imagine you're a Netflix addict and you just finished watching your favorite show, "Stranger Things." Now, you're looking for something new to watch, but you don't want to waste your time scrolling through endless lists of movies and shows that don't interest you.

Enter content-based filtering! This is when Netflix looks at the content of the shows and movies you've watched and recommends similar content based on what you seem to like.

It's kind of like having a personal Netflix genie who knows exactly what you want to watch next. Think of it like this: if you love sci-fi, Netflix will recommend other sci-fi shows and movies that you might enjoy. If you're a horror junkie, it'll recommend spooky flicks to give you the chills.

So, next time you're browsing Netflix and wondering what to watch next, remember that content-based filtering is working behind the scenes to bring you the shows and movies that are most likely to tickle your fancy! And if you're really lucky, maybe it'll even recommend a hidden gem that you never would have found on your own.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#removing stopwords
tfidf = TfidfVectorizer(stop_words='english')
```

```python
#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['Description'])
```

```python
#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

Here we are using **Cosine similarity**.
In data analysis, cosine similarity is a measure of similarity between two non-zero vectors defined in an inner product space. Cosine similarity is the cosine of the angle between the vectors; that is, it is the dot product of the vectors divided by the product of their lengths.

Given two n-dimensional vectors of attributes, A and B, the cosine similarity, cos(), is represented using a dot product and magnitude as
![](https://wikimedia.org/api/rest_v1/media/math/render/svg/0a4c9a778656537624a3303e646559a429868863)

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['Name of movie']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim= cosine_sim):
  idx= indices[title]

  # Get the pairwise similarity scores of all movies with that movie
  sim_scores= list(enumerate(cosine_sim[idx]))

  # Sort the movies based on the similarity scores
  sim_scores= sorted(sim_scores, key= lambda x:x[1], reverse= True)

  # Get the scores of the 10 most similar movies
  sim_scores= sim_scores[1:11]

  # Get the movie indices
  movie_indices= [i[0] for i in sim_scores]

  # Return the top 10 most similar movies
  return df['Name of movie'].iloc[movie_indices]
```

```python
features = ['Name of movie', 'Director', 'Star', 'genre', 'Description']

def create_soup(x):
    # Convert lists to strings
    star_str = ', '.join(x['Star']) if isinstance(x['Star'], list) else str(x['Star'])
    genre_str = ', '.join(x['Genre']) if isinstance(x['Genre'], list) else str(x['Genre'])

    # Concatenate all columns to create the soup
    return str(x['Name of movie']) + ' ' + str(x['Director']) + ' ' + star_str + ' ' + genre_str + ' ' + str(x['Description'])

# Apply the create_soup function to the DataFrame
filledna['soup'] = filledna.apply(create_soup, axis=1)
```

Now, instead of TFIDF Vectorizer, we will use Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count= CountVectorizer(stop_words= 'english')
count_matrix= count.fit_transform(filledna['soup'])

cosine_sim2= cosine_similarity(count_matrix, count_matrix)
```
